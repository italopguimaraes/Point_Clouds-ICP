## Importa as Bibliotecas e Define Funções Auxiliares Que Serão Utilizadas

```python
import numpy as np
import zipfile
import plotly.express as px
import shutil
from sklearn.neighbors import NearestNeighbors
import trimesh
import open3d as o3d
import transformations as transform
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import pandas as pd
import gdown
from pathlib import Path
import os
from colour import Color
%matplotlib widget

def create_dataset(url):
  '''Cria o dataset da nuvem de pontos, baixando os dados da url '''
  dest="Dataset/"
  #deleta o diretorio se ele ja existir, e cria um novo
  if os.path.isdir(dest):
    shutil.rmtree(dest)
  os.mkdir(dest)
  #muda o diretorio para dataset
  os.chdir('/content/'+dest)
  #Download Dataset
  output = 'KITTI-Sequence.zip'
  gdown.download(url, output, quiet=False)
  #Descompacta o Dataset
  with zipfile.ZipFile(output,"r") as zip_ref:
      zip_ref.extractall()
  # configura o diretorio dos dados de treino
  data_path = Path(dest)
  # Remove o arquivo dataset.zip
  os.remove(output)
  os.chdir('/content/')
  print('Download Dataset Point Clouds Concluído')
  return data_path

def calc_centroid_point_clouds(df):
  """
  Função para calcular o centróide de um point cloud, retornando as coordenadas X,Y,Z do centróide da nuven de pontos
  Entrada:
   df(pandas.DataFrame):DataFrame de entrada representando o point cloud
  Retorna:
    X(int):coordenada X do centróide
    Y(int):coordenada Y do centróide
    Z(int):coordenada Z do centróide
  """
  X_sum=0
  Y_sum=0
  Z_sum=0
  for i in range(len(df)):
    pc=df.iloc[i].values
    x=pc[0]
    y=pc[1]
    z=pc[2]
    X_sum+=x
    Y_sum+=y
    Z_sum+=z
  return float(X_sum)/len(df), float(Y_sum)/len(df), float(Z_sum)/len(df)

#calcula o ultimo ponto do GT
def calc_last_point_cloud(GT,pc):
  """
  Função para calcular o último/próximo ponto PB em coordenadas carteziandas, a partir do ponto imediatamente
  anterior PA também em coordenadas carteziandas e da matriz de transformação em coordenadas homogêneas M que ajusta
  perfeitamente os pontos PA e PB, através da multiplicação matricial entre M e PA, conforme a equação a seguir:
  PB=MxPA
  Entrada:
    GT:(numpy ndarray) matriz de transformação em coordenadas homogêneas 4x4 que mapeia o ponto PA em PB
    pc:(trimesh.base.Trimesh) nuvem de pontos PA anterior
  Retorna:
    pc_next:(trimesh.base.Trimesh) última/próxima nuvem de pontos PB calculada à partir da multiplicação entre M e PA
  """
  list_gt_fp=[]
  for line in np.array(pc.vertices):
    PX=line[0] #obtem os pontos das coordenadas x,y e z
    PY=line[1]
    PZ=line[2]
    #transforma em coordenadas homogeneas adicionando uma dimensao extra W=1
    new_line=np.array([PX,PY,PZ,1])
    #multiplica as matrizes M e PA para um ponto de dados
    last_point=np.dot(GT,new_line)
    list_gt_fp.append(last_point[:3]) #adiciona a lista

  list_gt_fp=np.array(list_gt_fp)
  pc_next=trimesh.base.Trimesh(list_gt_fp)
  return pc_next.convex_hull

def vizinho_mais_proximo(source, target):
    """
    Encontre o vizinho (euclidiano) mais próximo em dst para cada ponto em src
     Entrada:
         source: matriz de pontos fonte Nxm
         target: matriz de pontos destino Nxm
     Saída:
         distancias: distâncias euclidianas do vizinho mais próximo
         índices: índices dst do vizinho mais próximo
    """

    # assert source.shape == target.shape

    Neighbors = NearestNeighbors(n_neighbors=1, algorithm='kd_tree')
    Neighbors.fit(target)
    distancias, indices = Neighbors.kneighbors(source, return_distance=True)
    return distancias.ravel(), indices.ravel()

def melhor_transformacao_ajuste(A, B):
    """
    Calcula a transformação de melhor ajuste de mínimos quadrados que mapeia os pontos correspondentes A a B em m dimensões espaciais
     Entrada:
       A: matriz numpy Nxm de pontos correspondentes
       B: matriz numpy Nxm de pontos correspondentes
     Retorna:
       identity: (m+1)x(m+1) matriz de transformação homogênea que mapeia A em B
       dot: matriz de rotação mxm
       translate: vetor de tradução mx1
    """

    assert A.shape == B.shape

    # calcula o número de dimensões
    dim1 = A.shape[1]

    # translada pontos para seus centróides
    centroid_A = np.mean(A, axis=0)
    centroid_B = np.mean(B, axis=0)
    AA = A - centroid_A
    BB = B - centroid_B

    # rotaciona matrix
    M_rot = np.dot(AA.T, BB)
    Ut, St, Vt = np.linalg.svd(M_rot)
    dot = np.dot(Vt.T, Ut.T)

    # caso de reflexão especial
    if np.linalg.det(dot) < 0:
       Vt[dim1-1, :] *= -1
       dot = np.dot(Vt.T, Ut.T)

    # translação
    translate = centroid_B.T - np.dot(dot, centroid_A.T)

    # transformação homogênea
    identity = np.identity(dim1 + 1)
    identity[:dim1, :dim1] = dot
    identity[:dim1, dim1] = translate

    return identity, dot, translate

def melhor_transformacao_ajuste_point2plane(A, B, normals):
    """
    Calcula a transformação de melhor ajuste usando o metodo ponto a
    plano de mínimos quadrados que mapeia os pontos correspondentes A a B em m dimensões espaciais

    Input:
      A: matriz numpy Nx3 de pontos correspondentes
      B: matriz numpy Nx3 de pontos correspondentes
      normals: array numpy Nx3 de vetores normais de B
    Retorna:
      T: (m+1)x(m+1) matriz de transformação homogênea que mapeia A em B
      R: matriz de rotação mxm
      t: vetor de tradução mx1
    """

    assert A.shape == B.shape
    assert A.shape == normals.shape

    H = []
    b = []
    for i in range(A.shape[0]):
        dx = B[i, 0]
        dy = B[i, 1]
        dz = B[i, 2]
        nx = normals[i, 0]
        ny = normals[i, 1]
        nz = normals[i, 2]
        sx = A[i, 0]
        sy = A[i, 1]
        sz = A[i, 2]

        _a1 = (nz * sy) - (ny * sz)
        _a2 = (nx * sz) - (nz * sx)
        _a3 = (ny * sx) - (nx * sy)

        _a = np.array([_a1, _a2, _a3, nx, ny, nz])
        _b = (nx * dx) + (ny * dy) + (nz * dz) - (nx * sx) - (ny * sy) - (nz * sz)

        H.append(_a)
        b.append(_b)

    H = np.array(H)
    b = np.array(b)

    tr = np.dot(np.linalg.pinv(H), b)
    T = transform.euler_matrix(tr[0], tr[1], tr[2])
    T[0, 3] = tr[3]
    T[1, 3] = tr[4]
    T[2, 3] = tr[5]

    R = T[:3, :3]
    t = T[:3, 3]

    return T, R, t


def melhor_transformacao_ajuste_point2point(A, B):
    """
    Calcula a transformação de melhor ajuste usando o metodo ponto a
    ponto de mínimos quadrados que mapeia os pontos correspondentes A a B em m dimensões espaciais
    Entrada:
      A: matriz numpy Nxm de pontos correspondentes
      B: matriz numpy Nxm de pontos correspondentes
    Retorna:
      T: (m+1)x(m+1) matriz de transformação homogênea que mapeia A em B
      R: matriz de rotação mxm
      t: vetor de tradução mx1
    """

    assert A.shape == B.shape

    # calcula o numero de dimensões
    m = A.shape[1]

    # translada os pontos e seus centróides
    centroid_A = np.mean(A, axis=0)
    centroid_B = np.mean(B, axis=0)
    AA = A - centroid_A
    BB = B - centroid_B

    # rotaciona matrix
    H = np.dot(AA.T, BB)
    U, S, Vt = np.linalg.svd(H)
    R = np.dot(Vt.T, U.T)

    #caso de reflexão especial
    if np.linalg.det(R) < 0:
       Vt[m-1, :] *= -1
       R = np.dot(Vt.T, U.T)

    #translação
    t = centroid_B.T - np.dot(R, centroid_A.T)

    #transformação homogênea
    T = np.identity(m + 1)
    T[:m, :m] = R
    T[:m, m] = t

    return T, R, t

def point2pointICP(source_tm: "<class 'trimesh'>", target_tm: "<class 'trimesh'>",
        init_pose=None, num_iterations=10, tolerance=None, samplerate=1):
    """
    O método Iterativo Ponto mais próximo(ICP): encontra a transformação de melhor
    ajuste que mapeia os pontos da nuvem de origem source_tm para os pontos da nuvem
    de destino target_tm usando o método ponto para ponto
     Entrada:
         source_tm: matriz numpy Nxm de pontos de origem
         target_tm: matriz numpy Nxm do ponto de destino
         init_pose: (m+1)x(m+1) transformação homogênea
         num_iterations: numero de iterações, algoritmo termina após num_iterations
         tolerance: critérios de convergência
         samplerate: taxa de subamostragem
     Saída:
         Transformation_Matrix: transformação homogênea final que mapeia source_tm em target_tm
         MeanError: lista, que relata o erro médio da distância de cada iteração
    """

    # calcula os vértices e suas normais
    src_pts = np.array(source_tm.vertices)
    dst_pts = np.array(target_tm.vertices)
    src_pt_normals = np.array(source_tm.vertex_normals)
    dst_pt_normals = np.array(target_tm.vertex_normals)

    # sub-amostragem
    indices = np.random.uniform(0, 1, size=src_pts.shape[0])
    A = src_pts[indices < samplerate, :]
    A_normals = src_pt_normals[indices < samplerate, :]
    indices = np.random.uniform(0, 1, size=dst_pts.shape[0])
    B = dst_pts[indices < 1, :]
    B_normals = dst_pt_normals[indices < 1, :]

    # calcula o número de dimensões
    dim = A.shape[1]

    # torna os pontos homogêneos, copie-os para conservar os originais
    src = np.ones((dim+1, A.shape[0]))
    dst = np.ones((dim+1, B.shape[0]))
    src[:dim, :] = np.copy(A.T)
    dst[:dim, :] = np.copy(B.T)

    # aplica a estimativa de posição inicial
    if init_pose is not None:
        src = np.dot(init_pose, src)

    prev_error = 0
    MeanError = []

    for i in range(num_iterations):
        #encontra os vizinhos mais próximos entre os pontos de origem e de destino atuais
        distances, indices = vizinho_mais_proximo(src[:dim, :].T, dst[:dim, :].T)

        # combina cada ponto do conjunto de origem com o ponto mais próximo do conjunto de destino,
        matched_src_pts = src[:dim, :].T.copy()
        matched_dst_pts = dst[:dim, indices].T

        # calcula o ângulo entre as normais de 2 vértices correspondentes
        matched_src_pt_normals = A_normals.copy()
        matched_dst_pt_normals = B_normals[indices, :]
        angles = np.zeros(matched_src_pt_normals.shape[0])
        for k in range(matched_src_pt_normals.shape[0]):
            v1 = matched_src_pt_normals[k, :]
            v2 = matched_dst_pt_normals[k, :]
            cos_angle = v1.dot(v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
            angles[k] = np.arccos(cos_angle) / np.pi * 180

        # e rejeita qs correspondencias ruims
        dist_threshold = np.inf
        dist_bool_flag = (distances < dist_threshold)
        angle_threshold = 20
        angle_bool_flag = (angles < angle_threshold)
        reject_part_flag = dist_bool_flag * angle_bool_flag

        matched_src_pts = matched_src_pts[reject_part_flag, :]
        matched_dst_pts = matched_dst_pts[reject_part_flag, :]

        # calcula a transformação entre a origem atual e os pontos de destino mais próximos
        Transformation_Matrix, _, _ = melhor_transformacao_ajuste(matched_src_pts, matched_dst_pts)

        # atualiza a fonte atual
        src = np.dot(Transformation_Matrix, src)

        # imprimir iteração
        print('\ricp iteration: %d/%d ...' % (i+1, num_iterations), end='', flush=True)

        #verifica o erro
        mean_error = np.mean(distances[reject_part_flag])
        MeanError.append(mean_error)
        if tolerance is not None:
            if np.abs(prev_error - mean_error) < tolerance:
                print('\nbreak iteration, the distance between two adjacent iterations '
                      'is lower than tolerance (%.f < %f)'
                      % (np.abs(prev_error - mean_error), tolerance))
                break
        prev_error = mean_error

    #calcula a transformação final
    Transformation_Matrix, _, _ = melhor_transformacao_ajuste(A, src[:dim, :].T)
    print()

    return Transformation_Matrix, MeanError

def point2planeICP(source_tm: "<class 'trimesh'>", target_tm: "<class 'trimesh'>",
        init_pose=None, num_iterations=10, tolerance=None, samplerate=1):
    """
    O método Iterativo Ponto mais próximo(ICP): encontra a transformação de melhor
    ajuste que mapeia os pontos da nuvem de origem source_tm para os pontos da nuvem
    de destino target_tm usando o método ponto para plano
    Entrada:
         source_tm: matriz numpy Nxm de pontos de origem
         target_tm: matriz numpy Nxm do ponto de destino
         init_pose: (m+1)x(m+1) transformação homogênea
         num_iterations: numero de iterações, algoritmo termina após num_iterations
         tolerance: critérios de convergência
         samplerate: taxa de subamostragem
     Saída:
         Transformation_Matrix: transformação homogênea final que mapeia source_tm em target_tm
         MeanError: lista, que relata o erro médio da distância de cada iteração
    """

    # calcula os vertices e suas normais
    src_pts = np.array(source_tm.vertices)
    dst_pts = np.array(target_tm.vertices)
    src_pt_normals = np.array(source_tm.vertex_normals)
    dst_pt_normals = np.array(target_tm.vertex_normals)

    # aplica sub-amostragem
    indices = np.random.uniform(0, 1, size=src_pts.shape[0])
    A = src_pts[indices < samplerate, :]
    A_normals = src_pt_normals[indices < samplerate, :]
    indices = np.random.uniform(0, 1, size=dst_pts.shape[0])
    B = dst_pts[indices < samplerate, :]
    B_normals = dst_pt_normals[indices < samplerate, :]

    # calcula o numero de dimensões
    dim = A.shape[1]

    # torna os pontos homogêneos, e os copia para manter os originais
    src = np.ones((dim+1, A.shape[0]))
    dst = np.ones((dim+1, B.shape[0]))
    src[:dim, :] = np.copy(A.T)
    dst[:dim, :] = np.copy(B.T)

    # aplica a estimativa de posição inicial
    if init_pose is not None:
        src = np.dot(init_pose, src)

    prev_error = 0
    MeanError = []

    for i in range(num_iterations):
        #encontra os vizinhos mais próximos entre os pontos de origem e de destino atuais
        distances, indices = vizinho_mais_proximo(src[:dim, :].T, dst[:dim, :].T)

        # combina cada ponto do conjunto de origem com o ponto mais próximo do conjunto de destino,
        matched_src_pts = src[:dim, :].T.copy()
        matched_dst_pts = dst[:dim, indices].T

        # calcula o ângulo entre as normais de 2 vértices correspondentes
        matched_src_pt_normals = A_normals.copy()
        matched_dst_pt_normals = B_normals[indices, :]
        angles = np.zeros(matched_src_pt_normals.shape[0])
        for k in range(matched_src_pt_normals.shape[0]):
            v1 = matched_src_pt_normals[k, :]
            v2 = matched_dst_pt_normals[k, :]
            cos_angle = v1.dot(v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
            angles[k] = np.arccos(cos_angle) / np.pi * 180

        # e rejeita as correspondencias ruims
        dist_threshold = np.inf
        dist_bool_flag = (distances < dist_threshold)
        angle_threshold = 20
        angle_bool_flag = (angles < angle_threshold)
        reject_part_flag = dist_bool_flag * angle_bool_flag

        # obtém vértices correspondentes e normais dst_vertexes
        matched_src_pts = matched_src_pts[reject_part_flag, :]
        matched_dst_pts = matched_dst_pts[reject_part_flag, :]
        matched_dst_pt_normals = matched_dst_pt_normals[reject_part_flag, :]

        # calcula a transformação entre a matriz de origem atual e os pontos de destino mais próximos
        Transformation_Matrix, _, _ = melhor_transformacao_ajuste_point2plane(matched_src_pts, matched_dst_pts, matched_dst_pt_normals)

        # atualiza a fonte atual
        src = np.dot(Transformation_Matrix, src)

        # imprime iteração
        print('\ricp iteration: %d/%d ...' % (i+1, num_iterations), end='', flush=True)

        # verifica o erro
        mean_error = np.mean(distances[reject_part_flag])
        MeanError.append(mean_error)
        if tolerance is not None:
            if np.abs(prev_error - mean_error) < tolerance:
                print('\nbreak iteration, the distance between two adjacent iterations '
                      'is lower than tolerance (%.f < %f)'
                      % (np.abs(prev_error - mean_error), tolerance))
                break
        prev_error = mean_error

    # calcula a transformação final
    Transformation_Matrix, _, _ = melhor_transformacao_ajuste_point2point(A, src[:dim, :].T)
    print()

    return Transformation_Matrix, MeanError
```

# Procura os Melhores Parâmetros
### O Objetivo Nessa Etapa é Encontrar o Numero Ideal de Iterações e a Melhor Implemntação do Algorítimo ICP: ICP Point to Point ou ICP Point to Plane, Testando Várias Combinações de Números de Iterações e as Duas Implementações do ICP: Point to Point e Point to Plane, Calculando o Erro em Cada Iteração, e o Erro Total Acumulado no Final, Basicamente o Objetivo é Encontrar a Combinação de Parâmetros Que Produza Resultados Mais Próximos do Ground Truth, Escolhendo Aquela que Possúa o Menor Erro Total Acumulado Final Como Critério

```python
min_dist=float('inf')
best_algoritimo=None
best_iter=None

list_iter=[1,5,10,50,100,150,200,250,300,350,400]

for num_iter in list_iter:
  print('num_iter=',num_iter)
  len_error_point2point=0
  len_error_point2plane=0
  for i in range(len(list_points_GT)-1):
    source=list_points_GT[i]
    target=list_points_GT[i+1]
    H1, MeanErrors1 = point2pointICP(source, target, num_iterations=num_iter)
    H2, MeanErrors2 = point2planeICP(source, target, num_iterations=num_iter)
    len_error_point2point+=np.sum(abs(H1-Ground_Truth[i]))
    len_error_point2plane+=np.sum(abs(H2-Ground_Truth[i]))
  print('len_error_point2point=',len_error_point2point)
  print('len_error_point2plane=',len_error_point2plane)
  if len_error_point2point<=min_dist:
    min_dist=len_error_point2point
    best_algoritimo='point2point'
    best_iter=num_iter
  if len_error_point2plane<=min_dist:
    min_dist=len_error_point2plane
    best_algoritimo='point2plane'
    best_iter=num_iter
  print()
print('FINISH! FINAL RESULTS:')
print('BEST ALGORITHM:',best_algoritimo)
print('BEST ITERATION:',best_iter)
print('MIN DISTANCE:',min_dist)
```
